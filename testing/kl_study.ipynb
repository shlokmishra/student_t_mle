## KL divergence: Gibbs vs KDE as m varies

Investigates how unstable behavior (measured by KL(Gibbs || KDE)) appears/vanishes as m goes from 6 to 100 (k=1, Cauchy). Uses even m only (Gibbs x-update works in pairs).

```python
# Configuration
# Note: m must be even (Gibbs x-update works in pairs)
ms = [6, 10, 14, 20, 30, 50, 100]
k = 1.0  # Cauchy
mu_true = 2.0
T_gibbs = 50000
T_kde = 50000
base_params = {'mu_true': mu_true, 'prior_mean': 0.0, 'prior_std': 10.0, 'proposal_std_mu': 0.9, 'proposal_std_z': 0.03}
```

```python
# Run KL vs m study (takes a while)
kl_results = analysis.run_kl_vs_m_study(
    ms=ms, k=k, mu_true=mu_true,
    T_gibbs=T_gibbs, T_kde=T_kde, base_params=base_params,
    seed=0, verbose=True,
)
df_kl = pd.DataFrame(kl_results)
```

```python
# Plot KL vs m
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(df_kl['m'], df_kl['kl'], 'o-', linewidth=2, markersize=10)
axes[0].set_xlabel('m (sample size)')
axes[0].set_ylabel('KL(Gibbs || KDE)')
axes[0].set_title(f'KL divergence for k={k} (Cauchy)')
axes[0].grid(True, alpha=0.3)

axes[1].plot(df_kl['m'], df_kl['variance_ratio'], 's-', color='red', linewidth=2, markersize=10)
axes[1].axhline(1.0, color='black', linestyle='--', alpha=0.7)
axes[1].set_xlabel('m (sample size)')
axes[1].set_ylabel('Variance ratio (Gibbs/KDE)')
axes[1].set_title(f'Variance ratio for k={k}')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
print(df_kl.to_string(index=False))
```

## KL divergence for multiple k values

Same study but across k=1, 2, 3, 5. Shows that instability is primarily a Cauchy (k=1) issue.

```python
# Multi-k KL study
ks_multi = [1.0, 2.0, 3.0, 5.0]
ms_multi = [6, 10, 14, 20, 30, 50, 100]

kl_multi = analysis.run_kl_vs_m_multi_k(
    ks=ks_multi, ms=ms_multi, mu_true=mu_true,
    T_gibbs=T_gibbs, T_kde=T_kde, base_params=base_params,
    seed=0, verbose=True,
)
df_multi = pd.DataFrame(kl_multi)
```

```python
# Plot KL vs m for all k values
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for k_val in ks_multi:
    mask = df_multi['k'] == k_val
    axes[0].plot(df_multi[mask]['m'], df_multi[mask]['kl'], 'o-', linewidth=2, markersize=8, label=f'k={k_val}')

axes[0].set_xlabel('m (sample size)')
axes[0].set_ylabel('KL(Gibbs || KDE)')
axes[0].set_title('KL divergence vs m for different k')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].set_yscale('log')

for k_val in ks_multi:
    mask = df_multi['k'] == k_val
    axes[1].plot(df_multi[mask]['m'], df_multi[mask]['variance_ratio'], 'o-', linewidth=2, markersize=8, label=f'k={k_val}')

axes[1].axhline(1.0, color='black', linestyle='--', alpha=0.5)
axes[1].set_xlabel('m (sample size)')
axes[1].set_ylabel('Variance ratio (Gibbs/KDE)')
axes[1].set_title('Variance ratio vs m for different k')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].set_yscale('log')

plt.tight_layout()
plt.show()
```

```python
# Table: KL pivot by k and m
pivot_kl = df_multi.pivot(index='k', columns='m', values='kl')
print("KL(Gibbs || KDE) by k and m:")
print(pivot_kl.to_string(float_format='%.4f'))

print("\nVariance ratio (Gibbs/KDE) by k and m:")
pivot_vr = df_multi.pivot(index='k', columns='m', values='variance_ratio')
print(pivot_vr.to_string(float_format='%.4f'))
```
